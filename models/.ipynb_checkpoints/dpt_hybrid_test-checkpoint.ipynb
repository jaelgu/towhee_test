```python
# Import packages
import os
import glob
import torch
import cv2
#import argparse

import util.io

from torchvision.transforms import Compose

from dpt.models import DPTDepthModel, DPTSegmentationModel
#from dpt.midas_net import MidasNet_large
from dpt.transforms import Resize, NormalizeImage, PrepareForNet

#from util.misc import visualize_attention
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
mono_model_path = './weights/dpt_hybrid-midas-501f0c75.pt'
segment_model_path = './weights/dpt_hybrid-ade20k-53898607.pt'
```

```python
# Mono: dpt_hybrid

net_w = net_h = 384
mono_model = DPTDepthModel(
    path=mono_model_path,
    backbone="vitb_rn50_384",
    non_negative=True,
    enable_attention_hooks=False,
) 

mono_transform = Compose(
    [
        Resize(
            net_w,
            net_h,
            resize_target=None,
            keep_aspect_ratio=True,
            ensure_multiple_of=32,
            resize_method="minimal",
            image_interpolation_method=cv2.INTER_CUBIC,
        ),
        NormalizeImage(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
        PrepareForNet(),
    ]
)

mono_model.eval()
```

```python
# Segment: dpt_hybrid

net_w = net_h = 480
segment_model = DPTSegmentationModel(
    150,
    path=segment_model_path,
    backbone="vitb_rn50_384",
)

segment_transform = Compose(
    [
        Resize(
            net_w,
            net_h,
            resize_target=None,
            keep_aspect_ratio=True,
            ensure_multiple_of=32,
            resize_method="minimal",
            image_interpolation_method=cv2.INTER_CUBIC,
        ),
        NormalizeImage(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
        PrepareForNet(),
    ]
)

segment_model.eval()
```

```python
img_name = './input/test0.jpg'
img = util.io.read_image(img_name)
print(img.shape)

height, width, _ = img.shape
top = height - 352
left = (width - 1216) // 2
new_img = img[top : top + 352, left : left + 1216, :]
mono_img_input = mono_transform({"image": img})["image"]
segment_img_input = segment_transform({"image": img})["image"]
print(mono_img_input.shape, segment_img_input.shape)
```

```python
with torch.no_grad():
    # Mono prediction
    mono_sample = torch.from_numpy(mono_img_input).to(device).unsqueeze(0)
    mono_out = mono_model.forward(mono_sample)
    mono_prediction = torch.nn.functional.interpolate(
        mono_out.unsqueeze(1),
        size=img.shape[:2],
        mode="bicubic",
        align_corners=False,
    )
    mono_prediction = mono_prediction.squeeze().cpu().numpy()

    # Segment prediction:
    segment_sample = torch.from_numpy(segment_img_input).to(device).unsqueeze(0)
    segment_out = segment_model.forward(segment_sample)
    segment_prediction = torch.nn.functional.interpolate(
        segment_out,
        size=img.shape[:2],
        mode="bicubic",
        align_corners=False
    )
    segment_prediction = torch.argmax(segment_prediction, dim=1) + 1
    segment_prediction = segment_prediction.squeeze().cpu().numpy()
```

```python
filename_mono = './mg_test/mono_test'
filename_segment = './mg_test/segment_test'
util.io.write_depth(filename_mono, mono_prediction, bits=2, absolute_depth=False)
util.io.write_segm_img(filename_segment, img, segment_prediction, alpha=0.5)
```
